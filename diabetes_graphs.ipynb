```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
%matplotlib inline
# !conda install -c anaconda seaborn -y
```

```python
# Tableau graphs: https://public.tableau.com/profile/kylie6135#!/vizhome/DiabetesGraph1/Sheet1

diabetes_csv = "Resources/Diabetes.csv"
diabetes = pd.read_csv(diabetes_csv, encoding="utf-8")
```

```python
diabetes.head()
```

```python
diabetes_matrix = diabetes.corr()
diabetes_matrix
```

```python
import seaborn as sn
```

```python
all_data_plot = sn.heatmap(diabetes_matrix, annot=True, cmap="YlGnBu")
all_data_plot.set_title('All Data', fontsize=20)

plt.show()

# plt.savefig('all_data_graph.png')
```

```python
diabetes_clean = diabetes[(diabetes["Glucose"] > 0) & (diabetes["SkinThickness"] > 0) & (diabetes["Insulin"] > 0) & (diabetes["BMI"] > 0)]

diabetes_matrix_no_zero = diabetes_clean.corr()
clean_data_plot = sn.heatmap(diabetes_matrix_no_zero, annot=True)
clean_data_plot.set_title('Complete Data Only', fontsize=20)

plt.show()
# plt.savefig('complete_data_graph.png')
```
