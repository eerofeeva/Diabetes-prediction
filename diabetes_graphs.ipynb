```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# !conda install -c anaconda seaborn -y
```

```python
diabetes_csv = "Resources/Diabetes.csv"
diabetes = pd.read_csv(diabetes_csv, encoding="utf-8")
```

```python
diabetes.head()
```

```python
diabetes_matrix = diabetes.corr()
diabetes_matrix
```

```python
import seaborn as sn
```

```python
sn.heatmap(diabetes_matrix, annot=True)
plt.show()
```

```python
diabetes_outcome_1 = diabetes[diabetes["Outcome"] == 1]

diabetes_matrix_1 = diabetes_outcome_1.corr()
sn.heatmap(diabetes_matrix_1, annot=True)
plt.show()
```
